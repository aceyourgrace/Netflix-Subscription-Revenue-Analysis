```python
from nltk import word_tokenize
def tokenize(dataset):
    tk = word_tokenize(dataset)
    return tk
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
Lexicon = {}
for sentence in Req_String:
    tokens = tokenize(sentence)
    for token in tokens:
        sentiment_scores = sia.polarity_scores(token)
        Lexicon[token] = sentiment_scores
```

```python
def score_message(dataset):
    sentiment_score = sia.polarity_scores(dataset)
    compound_score = sentiment_score['compound']
    if compound_score >= 0.5:
        sentiment = 'Positive'
    elif compound_score <= -0.5:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'
    return(sentiment, compound_score)
```

```python
score_message(Req_String)
```
