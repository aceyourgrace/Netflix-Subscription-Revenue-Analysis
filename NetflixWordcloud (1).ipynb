```python
from nltk.tokenize import word_tokenize
import nltk
```

```python
Review_string = df1["Reviews"].to_string()
Review_string
```

```python
def tokenize(doc):
  return word_tokenize(doc)
tokens = tokenize(Review_string)
tokens
```

```python
# importing package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
```

```python
# setting stopwords and creating new functions for tokens that also remove the stopwords
stopwords = set(STOPWORDS)

stopwords.add('Netflix')

def count_frequenciesNEW(tokens):
    filtered_tokens = [token for token in tokens if token.lower() not in stopwords]
    nd = {token: filtered_tokens.count(token) for token in filtered_tokens}
    return nd
# count frequencies using the new function
frequencies_new = count_frequenciesNEW(tokens)
sorted_frequencies = sorted(frequencies_new.items(), key=lambda item: item[1], reverse=True)
sorted_frequencies
```

```python
# Assigning stopwords to variable named 'stopwords'

stopwords = set(STOPWORDS)

stopwords.add('Netflix')
stopwords.add('new')
stopwords.add('getting')
stopwords.add('movie')
stopwords.add('Im')
stopwords.add('Hi')

# Generating a word cloud

text_wc = WordCloud(

  background_color='white',

  max_words=2000,

  stopwords=stopwords)



text_wc.generate(Review_string)



plt.imshow(text_wc, interpolation='bilinear')

plt.axis('off')

plt.show()
```
